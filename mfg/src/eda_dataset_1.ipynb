```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset from CSV file
file_path = "../data/dataset_1.csv"  # Update this with the actual path of your CSV file
rop_data = pd.read_csv(file_path)
```

```python
# Check the structure of the dataset
print(rop_data.info())
```

```python
# Generate summary statistics for numerical columns
print(rop_data.describe())
```

```python
# Check for missing values
print("Missing Values:\n", rop_data.isnull().sum())
```

```python
# Define professional colors for each feature
colors = [
    "#1f77b4",  # Blue
    "#ff7f0e",  # Orange
    "#2ca02c",  # Green
    "#d62728",  # Red
    "#9467bd",  # Purple
    "#8c564b",  # Brown
    "#e377c2",  # Pink
    "#7f7f7f"   # Gray
]
```

```python
# Create histograms for each feature with a unique color
features = rop_data.columns
plt.figure(figsize=(15, 20))

for i, feature in enumerate(features):
    plt.subplot(4, 2, i + 1)  # Arrange plots in a grid (4 rows, 2 columns)
    plt.hist(rop_data[feature], bins=10, color=colors[i], edgecolor="black", alpha=0.8)
    plt.title(f"{feature}", fontsize=12)
    plt.xlabel(feature, fontsize=10)
    plt.ylabel("Frequency", fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.suptitle('Feature Distributions', fontsize=16, y=1.02)

# Save the plot as a JPG file
plt.savefig("../results/result_1/feature_distributions.jpg", format="jpg", dpi=300)

# Display the plot
plt.show()
```

```python
# Features to analyze
features = ['Depth', 'WOB', 'SURF_RPM', 'PHIF', 'VSH']

# Create scatter plots
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=rop_data[feature], y=rop_data['ROP_AVG'], color='blue')
    plt.title(f'Relationship between {feature} and ROP_AVG')
    plt.xlabel(feature)
    plt.ylabel('ROP_AVG')
    plt.grid(True)
    plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = rop_data.corr()

# Display correlation with the target variable (ROP_AVG)
print(correlation_matrix['ROP_AVG'])

# Visualize as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.savefig("../results/result_1/heat map.jpg", format="jpg", dpi=300)
plt.show()
```

```python
sns.pairplot(rop_data, x_vars=features, y_vars='ROP_AVG', height=5, aspect=1, kind='scatter')
plt.suptitle('Pairwise Relationships with ROP_AVG', y=1.02)
plt.show()
```

```python
from scipy.stats import pearsonr

# Compute Pearson correlation for each feature
for feature in features:
    corr, p_value = pearsonr(rop_data[feature], rop_data['ROP_AVG'])
    print(f'Feature: {feature}, Pearson Correlation: {corr:.2f}, P-value: {p_value:.3f}')
```

```python
from scipy.stats import spearmanr

# Compute Spearman correlation for each feature
for feature in features:
    corr, p_value = spearmanr(rop_data[feature], rop_data['ROP_AVG'])
    print(f'Feature: {feature}, Spearman Correlation: {corr:.2f}, P-value: {p_value:.3f}')
```
